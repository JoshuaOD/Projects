# Data Cleaning NBA Data

```python
import pandas as pd
```

```python
nba_data = pd.read_csv('nba_player_stats_1992-2022.csv')
```

```python
nba_data
```

```python
nba_data.shape
```

```python
del nba_data['Unnamed: 0']
del nba_data['Rk']
```

```python
nba_data.isnull().sum()
```

```python
player_names = nba_data["Player"]
nba_data[player_names.isin(player_names[player_names.duplicated()])].sort_values("Player")
```

```python
def single_team(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"]=="TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

nba_data = nba_data.groupby(["Player", "Year"]).apply(single_team)
```

```python
nba_data
```

```python
nba_data.index = nba_data.index.droplevel()
```
