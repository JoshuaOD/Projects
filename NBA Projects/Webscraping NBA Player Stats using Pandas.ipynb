```python
import requests
import pandas as pd
```

```python
years = list(range(2022, 1991,-1))
len(years)
```

```python
url = 'https://www.basketball-reference.com/leagues/NBA_1992_totals.html'
data = requests.get(url)
```

```python
matches = pd.read_html(data.text, match='Player Totals Table')[0]
```

```python
matches
```

```python
player_stats = list()

for year in years:
    urls = (f'https://www.basketball-reference.com/leagues/NBA_{year}_totals.html')
    data = requests.get(urls)
    player_stat = pd.read_html(data.text, match='Player Totals Table')[0]
    player_stat['Year'] = year 
    player_stats.append(player_stat)
```

```python
nba_player_stats_1992_2022 = pd.concat(player_stats)
```
